```python
comb_dic = dict()
X_train_category = dict()
for i in comb:
    comb_dic[i]=0
    X_train_category[i]=[]
    
for i in range(153):
    comb_dic[(i,i)]=0
    X_train_category[(i,i)]=[]
```

```python

for i in X_train:
    k = id_to_category[i[0]]
    v = id_to_category[i[1]]
    if k>=v:
        comb_dic[(v,k)] += 1
        X_train_category[(v,k)] = i
    else:
        comb_dic[(k,v)] += 1
        X_train_category[(k,v)] = i

        
```
